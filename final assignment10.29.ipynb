```python
#Which vendors/plants are causing the greatest defect quantity?
```

```python
# Group the data using 'Vendor' and calculate the total defect quantity for each vendor.
vendor_total_defect = df_suppliers.groupby('Vendor')['Total Defect Qty'].sum()

# Group the data using 'Plant Location' and calculate the total defect quantity for each plant location.
plant_total_defect = df_suppliers.groupby('Plant Location')['Total Defect Qty'].sum()

# (Optional) Sort the dictionaries with the summed values in descending order for curiosity.
sorted_vendor_total_defect = vendor_total_defect.sort_values(ascending=False)
sorted_plant_total_defect = plant_total_defect.sort_values(ascending=False)

# Find the vendor with the highest total defect quantity, indicating the one causing the greatest defects among all the vendors.
vendor_defect_greatest = vendor_total_defect.idxmax()

# Find the plant location with the highest total defect quantity, indicating the one causing the greatest defects among all the plant locations.
plant_defect_greatest = plant_total_defect.idxmax()

# Print the results to identify the vendor and plant location with the greatest defect quantity.
print(f'{vendor_defect_greatest} is responsible for the highest defect quantity among all the vendors.')
print(f'{plant_defect_greatest} is responsible for the highest defect quantity among all the plant locations.')
```

```python
#Which vendors/plants are causing the greatest downtime?
```

```python
# Group the data using 'Vendor' and calculate the total downtime minutes for each vendor.
vendor_total_downtime = df_suppliers.groupby('Vendor')['Total Downtime Minutes'].sum()

# Group the data using 'Plant Location' and calculate the total downtime minutes for each plant location.
plant_total_downtime = df_suppliers.groupby('Plant Location')['Total Downtime Minutes'].sum()

# (Optional) Sort the dictionaries with the summed values in descending order for curiosity.
sorted_vendor_total_downtime = vendor_total_downtime.sort_values(ascending=False)
sorted_plant_total_downtime = plant_total_downtime.sort_values(ascending=False)

# Find the vendor with the highest total downtime causing the greatest defect quantity.
vendor_downtime_greatest = vendor_total_downtime.idxmax()

# Find the plant location with the highest total downtime causing the greatest defect quantity.
plant_downtime_greatest = plant_total_downtime.idxmax()

# Print the results indicating the vendor and plant location with the greatest defect quantity.
print(f'{vendor_downtime_greatest} is responsible for the highest defect quantity among all the vendors.')
print(f'{plant_downtime_greatest} is responsible for the highest defect quantity among all the plant locations.')
```

```python
#Is there a particular combination of material and vendor that perform poorly?
```

```python
# Create a result dataframe
result_vm_df = pd.DataFrame()
data = {'Vendor': [],
        'Material Type': [],
        'Portion(%)':[]}
result_vm_df = pd.DataFrame(data)

# define the function of appending results
def df_result_create(vendor, material_type, portion):
    # Convert the portion to a percentage
    round_portion = portion.round(4)
    percentage_portion = round_portion * 100
    # Add result row to the result_df
    result_vm_row = {'Vendor': vendor, 'Material Type': material_type, 'Portion(%)': percentage_portion}   
    result_vm_df.loc[len(result_vm_df)] = result_vm_row

# Get unique material types
material_type = df_suppliers['Material Type'].unique()
for m in material_type:
    # Find all m-material associated rows and count the number of vendors
    m_row = df_suppliers.loc[df_suppliers['Material Type'] == m]
    list_vendor = m_row['Vendor'].unique()
    # Threshold
    nb_vendor = len(list_vendor)
    threshold = 3 * (1 / nb_vendor)# We define "poorly perform" as 3 times of average defect portion of each vendor for m-material  
    # Sum up all m-material defect qty
    m_qty = m_row['Total Defect Qty'].sum() 
    for v in list_vendor:
        # Find all v-vendor associated rows for m-material and count the number of defect
        v_m_row = m_row.loc[m_row['Vendor'] == v]
        v_m_qty = v_m_row['Total Defect Qty'].sum()
        # Calculate the defect portion of this material caused by v-vendor
        v_m_portion = v_m_qty / m_qty
        # Compare the portion to the threshold
        if v_m_portion > threshold:
            df_result_create(v, m, v_m_portion)

result_vm_df
```

```python
# Create a result dataframe
result_vp_df = pd.DataFrame()
data = {'Vendor': [],
        'Plant Location': [],
        'Portion(%)':[]}
result_vp_df = pd.DataFrame(data)

# define the function of appending results
def df_result_create(vendor, plant, portion):
    # Convert the portion to a percentage
    round_portion = portion.round(4)
    percentage_portion = round_portion * 100
    # Add result row to the result_df
    result_vp_row = {'Vendor': vendor, 'Plant Location': plant, 'Portion(%)': percentage_portion}   
    result_vp_df.loc[len(result_vp_df)] = result_vp_row

# Get unique Plant Locations
plant = df_suppliers['Plant Location'].unique()
for p in plant:
    # Find all plants associated rows and count the number of vendors
    p_row = df_suppliers.loc[df_suppliers['Plant Location'] == p]
    list_vendor = p_row['Vendor'].unique()
    # Threshold
    nb_vendor = len(list_vendor)
    threshold = 3 * (1 / nb_vendor)# We define "poorly perform" as 3 times of average defect portion of each vendor for each plant  
    # Sum up all plant defect qty
    p_qty = p_row['Total Defect Qty'].sum() 
    for v in list_vendor:
        # Find all v-vendor associated rows for p-plant and count the number of defect
        v_p_row = p_row.loc[p_row['Vendor'] == v]
        v_p_qty = v_p_row['Total Defect Qty'].sum()
        # Calculate the defect portion of this plant caused by v-vendor
        v_p_portion = v_p_qty / p_qty
        # Compare the portion to the threshold
        if v_p_portion > threshold:
            df_result_create(v, p, v_p_portion)

result_vp_df
```

```python
!pip install XlsxWriter

# Create a result dataframe
df_result = pd.DataFrame()
data = {'Vendor': [],
        'Material Type':[],
        'Plant Location': [],
        'Total Defect Qty':[],
        'Total Downtime Minutes':[]}
df_result = pd.DataFrame(data)

# define the function of appending results
def df_result_create(vendor, material, plant, total_defect_qty, total_downtime_minutes):
    # Add result row to the result_df
    result_row = {'Vendor': vendor, 'Material Type': material, 'Plant Location': plant, 'Total Defect Qty': total_defect_qty, 'Total Downtime Minutes': total_downtime_minutes}   
    df_result.loc[len(df_result)] = result_row


# Get unique material types
material_type = df_suppliers['Material Type'].unique()
for m in material_type:
    # Find all m-material associated rows and count the number of vendors
    m_row = df_suppliers.loc[df_suppliers['Material Type'] == m]
    list_vendor = m_row['Vendor'].unique()
    for v in list_vendor:
        # Find all v-vendor associated rows for m-material and count the number of defect
        v_m_row = m_row.loc[m_row['Vendor'] == v]
        list_plant = v_m_row['Plant Location'].unique()
        nb_plant = len(list_plant)
        if nb_plant > 1:
            for p in list_plant:
                defect = v_m_row.loc[v_m_row['Plant Location'] == p]['Total Defect Qty']
                downtime = v_m_row.loc[v_m_row['Plant Location'] == p]['Total Downtime Minutes']
                df_result_create(v, m, p, defect.sum(), downtime.sum())
df_result

#df_result.to_excel('output_data.xlsx', index=False)


                
```
